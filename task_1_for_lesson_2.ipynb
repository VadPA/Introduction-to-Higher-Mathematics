```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

### Функция расчета суммы двух векторов

```python
def sum_vectors(x, y):
    try:
        return [x[i] + y[i] for i in range(len(x))]
    except Exception as e:
        print(f'Произошла ошибка: {e}')
```

### Функция расчета длины вектора

```python
def length_vectors(x):
    sum_of_squares = 0
    for i in x:
        sum_of_squares += i * i
    return math.sqrt(sum_of_squares)
```

```python
x1 = [10, 10, 10]
x2 = [0, 0, -10]
y = sum_vectors(x1, x2)
print(f'Сумма векторов {x1} и {x2}: {sum_vectors(x1, x2)}')
print(f'Длина вектора {x1}: {length_vectors(x1):0.2f}')
print(f'Длина вектора {x2}: {length_vectors(x2):0.2f}')
```

### Функция для расчета точек окружности

```python
def f_circle(x, R):
    return (math.sqrt((R)**2-(x)**2))
```

### Функция для расчета точек эллипса

```python
def f_elips(x, a, b):
    return (math.sqrt((a)**2-(x)**2)*b/a)
```

### Функция для расчета точек гиперболы

```python
def f_giper(x, a, b):
    return (math.sqrt((a)**2+(x)**2)*b/a)
```

```python
y1 = []
y2 = []
y3 = []
y4 = []
# координаты центра и радиус окружности
x_ci = -7  
y_ci = -3
R = 4
# координаты центра элипса
x_el = 3  
y_el = 5
a = 4
b = 2

x = [el * 0.01 for el in range(-100*R, R*100 + 1)]
# print(x)
for el in x:
    y1.append(f_circle(el, R)+y_ci)
    y2.append(-f_circle(el, R)+y_ci)
    y3.append(f_elips(el, a, b)+y_el)
    y4.append(-f_elips(el, a, b)+y_el)

x = [x_ci + el * 0.01 for el in range(-100*R, R*100 + 1)]
plt.plot(x,y1)
plt.plot(x,y2)
x = [x_el + el * 0.01 for el in range(-100*R, R*100 + 1)]
plt.plot(x,y3)
plt.plot(x,y4)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

### Построение двух параллельных плоскостей

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = (2*X - 3*Y + 12)/3
Z1 = (2*X - 3*Y)/3
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z1)
#ax.plot_wireframe(X, Y, Z)
ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```
