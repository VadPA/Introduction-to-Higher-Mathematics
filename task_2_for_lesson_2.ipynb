### Функция перевода в полярные координаты

```python
def polarn_XY(ro, fi):
    return (ro * np.cos(fi), ro * np.sin(fi))
```

### Рисуем окружность в полярных координатах (и ещё пару фигур)

```python
a = np.linspace(0, 2*np.pi, 801)
R = 5
ro = R
x, y = polarn_XY(ro, a)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-7, 6.0)
plt.grid(True)
plt.show()
```

```python
a = np.linspace(0, 2*np.pi, 801)
ro = 1 - np.sin(a)
ro1 = np.sin(3*a)
x, y = polarn_XY(ro, a)
x1, y1 = polarn_XY(ro1, a)
plt.plot(x, y)
plt.plot(x1, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2.5, 1.0)
plt.grid(True)
plt.show()
```

```python
a = np.linspace(0, 8*np.pi, 801)
b = 35
k = 6
te = 0
ro = b * (np.cos(a*k+te))
x, y = polarn_XY(ro, a)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-65, 55)
plt.grid(True)
plt.show()
```

### График прямой линии в полярных координатах

```python
a = np.linspace(1*np.pi/4, 1*np.pi/4, 1)
R = np.linspace(-10, 10, 100)
ro = R
x, y = polarn_XY(ro, a)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-7, 6.0)
plt.grid(True)
plt.show()
```

### Решить систему уравнений

```python
x= np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, (x**2 - 1))
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-7, 8.0)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def f(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

x1, y1 = fsolve(f, (-2, 2))

print(x1, y1)
```
