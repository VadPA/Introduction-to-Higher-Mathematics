#### Рисую график функции y = k * np.cos(x-a) + b , при нескольких параметрах

```python
x = np.linspace(-2*np.pi, 3*np.pi, 201)
k = 0.5
a = np.pi/6
b = 2
y = k * np.cos(x-a) + b 
k = 0.8
a = np.pi/3
b = 3
y1 = k * np.cos(x-a) + b 
plt.plot(x, y)
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

### Функция перевода в полярные координаты

```python
def polarn_XY(ro, fi):
    return (ro * np.cos(fi), ro * np.sin(fi))
```

### Рисуем окружность в полярных координатах (и ещё пару фигур)

```python
a = np.linspace(0, 2*np.pi, 801)
R = 5
ro = R
x, y = polarn_XY(ro, a)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-7, 6.0)
plt.grid(True)
plt.show()
```

```python
a = np.linspace(0, 2*np.pi, 801)
ro = 1 - np.sin(a)
ro1 = np.sin(3*a)
x, y = polarn_XY(ro, a)
x1, y1 = polarn_XY(ro1, a)
plt.plot(x, y)
plt.plot(x1, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2.5, 1.0)
plt.grid(True)
plt.show()
```

```python
a = np.linspace(0, 8*np.pi, 801)
b = 35
k = 6
te = 0
ro = b * (np.cos(a*k+te))
x, y = polarn_XY(ro, a)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-65, 55)
plt.grid(True)
plt.show()
```

### График прямой линии в полярных координатах

```python
a = np.linspace(1*np.pi/4, 1*np.pi/4, 1)
R = np.linspace(-10, 10, 100)
ro = R
x, y = polarn_XY(ro, a)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-7, 6.0)
plt.grid(True)
plt.show()
```

### Решить систему уравнений

<img src="System1.png">

```python
x= np.linspace(-6, 6, 201)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, (x**2 - 1))
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-7, 20.0)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def f(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

x1, y1 = fsolve(f, (-2, 2))
x2, y2 = fsolve(f, (3, 5))
x3, y3 = fsolve(f, (4, 20))

print(f'{x1:0.2f}, {y1:0.2f}')
print(f' {x2:0.2f}, {y2:0.2f}')
print(f' {x3:0.2f}, {y3:0.2f}')
```

Упущено одно решение: x = 0 y = -1 (0,-1)

### Ответ: решением системы будут пары (x;y): (-1.58;1.5), (2.62;5.85), (4.2;16.64), (0;-1)

### Подставляя в неравенство значение y из уравнения, то исходную систему можно свести к решению следующего неравенства:
