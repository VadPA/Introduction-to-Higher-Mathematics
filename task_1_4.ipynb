```python
%matplotlib notebook
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
k1 = 1
k2 = 2
x=np.arange(-math.pi *2, math.pi *2, 0.05)
fig=plt.figure()
plt.plot(x, np.sin(k1 * x))
plt.plot(x, np.sin(k2 * x))
```

```python
x=np.arange(-math.pi * 1, math.pi * 1, 0.05)
fig=plt.figure()
for i in range(0,5):
    plt.plot(x, np.sin(i * x))
```
